PRIORIZAR LA HIPOTESIS:

```python
# Calculamos el ICE score
hypotheses['ICE'] = (hypotheses['impact'] * hypotheses['confidence']) / hypotheses['effort']

# Ordenamos en orden descendente de ICE
hypotheses_ice = hypotheses.sort_values(by='ICE', ascending=False)

# Mostramos los resultados ordenados
display(hypotheses_ice[['hypothesis', 'ICE']])
```

```python
# Calculamos el RICE score
hypotheses['RICE'] = (hypotheses['reach'] * hypotheses['impact'] * hypotheses['confidence']) / hypotheses['effort']

# Ordenamos en orden descendente de RICE
hypotheses_rice = hypotheses.sort_values(by='RICE', ascending=False)

# Mostramos los resultados ordenados
display(hypotheses_rice[['hypothesis', 'RICE']])
```

```python
# üìä Comparar ICE vs RICE
comparison = hypotheses[['hypothesis', 'ICE', 'RICE']].copy()

# Ordenamos por ICE
ice_sorted = comparison.sort_values(by='ICE', ascending=False)
print("üîπ Orden por ICE:")
display(ice_sorted)

# Ordenamos por RICE
rice_sorted = comparison.sort_values(by='RICE', ascending=False)
print("üîπ Orden por RICE:")
display(rice_sorted)
```

                         COMPARACION Y EXPLICACION:
	
    Con ICE, se priorizan las hip√≥tesis que tienen mayor impacto y confianza respecto al esfuerzo.
	Con RICE, adem√°s de esas variables, se incluye el alcance (Reach), es decir, cu√°ntos usuarios se ver√≠an afectados.

Esto hace que:
	‚Ä¢	Hip√≥tesis como ‚ÄúAdd a subscription form‚Äù, aunque tenga un ICE medio, sube al primer lugar en RICE porque afecta a un gran n√∫mero de usuarios.
	‚Ä¢	Por el contrario, ‚ÄúLaunch a promotion‚Äù, aunque ten√≠a un ICE alto, baja de prioridad porque su alcance es menor.

El m√©todo RICE es m√°s completo porque considera cu√°ntos usuarios se ver√°n beneficiados, no solo la confianza e impacto.
En este caso, el negocio deber√≠a priorizar la hip√≥tesis del formulario de suscripci√≥n (RICE m√°s alto), ya que tiene gran alcance, buen impacto y requiere esfuerzo razonable.

```python
plt.figure(figsize=(12, 6))

# Posiciones
x = np.arange(len(hypotheses['hypothesis']))

# Gr√°fico de barras comparativo
plt.bar(x - 0.2, hypotheses['ICE'], width=0.4, label='ICE', color='skyblue')
plt.bar(x + 0.2, hypotheses['RICE'], width=0.4, label='RICE', color='orange')

# Etiquetas de eje X con nombres reales
plt.xticks(x, hypotheses['hypothesis'], rotation=45, ha='right')

# T√≠tulos y leyenda
plt.title("Comparaci√≥n de priorizaci√≥n: ICE vs RICE", fontsize=14, fontweight='bold')
plt.xlabel("Hip√≥tesis", fontsize=12)
plt.ylabel("Puntuaci√≥n", fontsize=12)
plt.legend()

plt.tight_layout()
plt.show()
```

El gr√°fico compara las puntuaciones ICE y RICE para cada hip√≥tesis. Se observa que el m√©todo RICE cambia la priorizaci√≥n al considerar el alcance (Reach), lo que permite identificar ideas con mayor potencial de impacto en usuarios, incluso si su facilidad o confianza es menor.

RICE ofrece una visi√≥n m√°s completa de la prioridad, ya que pondera cu√°ntos usuarios se ver√°n afectados por cada cambio.
üöÄ Las hip√≥tesis con alto alcance y buen impacto suben en la lista de prioridad, mientras que las de menor alcance bajan.

ANALISIS DEL TEST A/B:

```python
revenue_daily = orders.groupby(['date', 'group'])['revenue'].sum().reset_index()
revenue_daily['cumulative_revenue'] = revenue_daily.groupby('group')['revenue'].cumsum()
```

```python
plt.figure(figsize=(10,6))

# Filtramos por grupo A y B
group_a = revenue_daily[revenue_daily['group'] == 'A']
group_b = revenue_daily[revenue_daily['group'] == 'B']

# Graficamos
plt.plot(group_a['date'], group_a['cumulative_revenue'], label='Grupo A', marker='o')
plt.plot(group_b['date'], group_b['cumulative_revenue'], label='Grupo B', marker='o')
plt.title('Ingreso acumulado por grupo')
plt.xlabel('Fecha')
plt.ylabel('Ingreso acumulado')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

El grupo B muestra un ingreso acumulado mayor que el grupo A, lo que indica que la modificaci√≥n aplicada al grupo B podr√≠a estar generando un impacto positivo en las ventas.
Sin embargo, es necesario seguir analizando otras m√©tricas (como el tama√±o promedio del pedido y la conversi√≥n) para confirmar si la diferencia es significativa.

```python
daily_revenue = orders.groupby(['date', 'group'])['revenue'].sum().reset_index()
daily_orders_count = orders.groupby(['date', 'group'])['transactionid'].count().reset_index()
daily_orders_count.rename(columns={'transactionid': 'orders_count'}, inplace=True)
daily_data = pd.merge(daily_revenue, daily_orders_count, on=['date', 'group'])


daily_data['cum_revenue'] = daily_data.groupby('group')['revenue'].cumsum()
daily_data['cum_orders'] = daily_data.groupby('group')['orders_count'].cumsum()

daily_data['avg_order_value'] = daily_data['cum_revenue'] / daily_data['cum_orders']
```

```python
plt.figure(figsize=(10,6))
for group in daily_data['group'].unique():
    group_data = daily_data[daily_data['group'] == group]
    plt.plot(group_data['date'], group_data['avg_order_value'], label=f'Grupo {group}')

plt.title('Tama√±o de pedido promedio acumulado por grupo')
plt.xlabel('Fecha')
plt.ylabel('Tama√±o promedio del pedido ($)')
plt.legend()
plt.grid(True)
plt.show()
```

El grupo B presenta un tama√±o de pedido promedio acumulado m√°s alto durante la mayor parte del experimento, especialmente en la mitad del per√≠odo.
Esto sugiere que la modificaci√≥n aplicada al grupo B podr√≠a estar influyendo positivamente en el monto promedio por pedido.
Sin embargo, las fluctuaciones y disminuci√≥n final indican que el efecto no es estable, por lo que ser√≠a necesario verificar si estas diferencias son estad√≠sticamente significativas

```python
groupA = daily_data[daily_data['group'] == 'A'][['date', 'avg_order_value']]
groupB = daily_data[daily_data['group'] == 'B'][['date', 'avg_order_value']]


groupA = groupA.rename(columns={'avg_order_value': 'avg_order_value_A'})
groupB = groupB.rename(columns={'avg_order_value': 'avg_order_value_B'})


merged = pd.merge(groupA, groupB, on='date')


merged['relative_difference'] = (merged['avg_order_value_B'] / merged['avg_order_value_A']) - 1
```

```python
plt.figure(figsize=(10,6))
plt.plot(merged['date'], merged['relative_difference'], marker='o', color='purple')
plt.axhline(y=0, color='gray', linestyle='--')
plt.title('Diferencia relativa en el tama√±o promedio del pedido (B vs A)')
plt.xlabel('Fecha')
plt.ylabel('Diferencia relativa')
plt.grid(True)
plt.show()

display(merged.head())
```

El grupo B mostr√≥ un tama√±o promedio de pedido mayor que el grupo A durante la mayor parte del experimento, con picos de hasta +30 %.
Aunque hubo fluctuaciones al inicio, la diferencia se estabiliz√≥ en valores positivos, indicando una ventaja consistente del grupo B.
Es necesario confirmar si esta diferencia es estad√≠sticamente significativa antes de sacar conclusiones finales.

```python
orders_daily = orders.groupby(['date', 'group'])['transactionid'].nunique().reset_index()
orders_daily.rename(columns={'transactionid': 'orders_per_day'}, inplace=True)
```

```python
visits_daily = visits.groupby(['date', 'group'])['visits'].sum().reset_index()
conversion = pd.merge(orders_daily, visits_daily, on=['date', 'group'])
conversion['conversion_rate'] = conversion['orders_per_day'] / conversion['visits']
```

```python
plt.figure(figsize=(12,6))

for group in ['A', 'B']:
    group_data = conversion[conversion['group'] == group]
    plt.plot(group_data['date'], group_data['conversion_rate'], label=f'Grupo {group}')

plt.title('Tasa de conversi√≥n diaria por grupo')
plt.xlabel('Fecha')
plt.ylabel('Tasa de conversi√≥n (pedidos / visitas)')
plt.legend()
plt.grid(True)
plt.show()
```

Ambos grupos muestran fluctuaciones considerables a lo largo del tiempo. El grupo B tiene una tasa de conversi√≥n ligeramente superior en varios d√≠as, pero sin una tendencia clara y constante frente al grupo A.
Por lo tanto, no hay evidencia suficiente para afirmar una ventaja significativa de un grupo sobre el otro solo con esta gr√°fica; se requiere una prueba estad√≠stica para confirmar si la diferencia es significativa.

```python
orders_per_user = orders.groupby('visitorid')['transactionid'].nunique().reset_index()

orders_per_user.columns = ['visitorid', 'orders_count']

plt.figure(figsize=(10, 6))
plt.scatter(orders_per_user.index, orders_per_user['orders_count'], alpha=0.5)
plt.title('Gr√°fico de dispersi√≥n: n√∫mero de pedidos por usuario')
plt.xlabel('Usuarios')
plt.ylabel('N√∫mero de pedidos')
plt.grid(True)
plt.show()
```

La mayor√≠a de los usuarios realizaron solo 1 pedido, aunque se observan algunos casos aislados con varios pedidos, que podr√≠an ser anomal√≠as o usuarios altamente activos.
Estos valores at√≠picos deben revisarse con los percentiles 95 y 99 en el siguiente paso para definir l√≠mites de normalidad.

```python
percentil_95 = np.percentile(orders_per_user['orders_count'], 95)
percentil_99 = np.percentile(orders_per_user['orders_count'], 99)

print(f"Percentil 95: {percentil_95}")
print(f"Percentil 99: {percentil_99}")
```

```python

percentil_95_revenue = np.percentile(orders['revenue'], 95)
percentil_99_revenue = np.percentile(orders['revenue'], 99)

print(f"Percentil 95 del precio: {percentil_95_revenue}")
print(f"Percentil 99 del precio: {percentil_99_revenue}")
```

El 95 % de los pedidos tienen un precio igual o menor a 435.54, y el 99 % tienen un precio igual o menor a 900.90.
Por lo tanto, cualquier pedido con un precio mayor a 900.90 puede considerarse una anomal√≠a (outlier), ya que representa valores at√≠picos que podr√≠an distorsionar el an√°lisis de ingresos y promedios.

Estos casos deben ser analizados por separado o filtrados antes de realizar pruebas estad√≠sticas.


```python
visitors_A = visits[visits['group'] == 'A']
visitors_B = visits[visits['group'] == 'B']

orders_A = orders[orders['group'] == 'A']['visitorid'].unique()
orders_B = orders[orders['group'] == 'B']['visitorid'].unique()

sample_A = visitors_A['visits'].apply(lambda x: 1 if x in orders_A else 0)
sample_B = visitors_B['visits'].apply(lambda x: 1 if x in orders_B else 0)


stat, p_value = stats.mannwhitneyu(sample_A, sample_B)
print('Estad√≠stico:', stat)
print('p-valor:', p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("‚úÖ Hay una diferencia estad√≠sticamente significativa en la conversi√≥n entre los grupos.")
else:
    print("‚ùå No hay una diferencia estad√≠sticamente significativa en la conversi√≥n entre los grupos.")
```

La tasa de conversi√≥n del grupo B es ligeramente superior a la del grupo A, pero con la diferencia no es estad√≠sticamente significativa.
Esto significa que no hay evidencia suficiente para afirmar que el cambio implementado en el grupo B haya tenido un impacto real en la conversi√≥n de usuarios.

```python
revenue_A = orders[orders['group'] == 'A']['revenue']
revenue_B = orders[orders['group'] == 'B']['revenue']

stat, p_value = stats.mannwhitneyu(revenue_A, revenue_B)

print("üìä Estad√≠stico:", stat)
print("üìâ p-valor:", p_value)

alpha = 0.05
if p_value < alpha:
    print("‚úÖ Hay una diferencia estad√≠sticamente significativa en el tama√±o promedio del pedido entre los grupos.")
else:
    print("‚ùå No hay una diferencia estad√≠sticamente significativa en el tama√±o promedio del pedido entre los grupos.")
```

```python
orders_filtered = orders[
    (orders['revenue'] <= 900.9) &
    (orders['revenue'] > 0)
]
orders_A_filtered = orders_filtered[orders_filtered['group'] == 'A']['visitorid'].unique()
orders_B_filtered = orders_filtered[orders_filtered['group'] == 'B']['visitorid'].unique()


sample_A_filtered = visitors_A['visits'].apply(lambda x: 1 if x in orders_A_filtered else 0)
sample_B_filtered = visitors_B['visits'].apply(lambda x: 1 if x in orders_B_filtered else 0)


stat_filtered, p_value_filtered = stats.mannwhitneyu(sample_A_filtered, sample_B_filtered)

print('üìä Estad√≠stico:', stat_filtered)
print('üß™ p-valor:', p_value_filtered)


alpha = 0.05

if p_value_filtered < alpha:
    print("‚úÖ Hay una diferencia estad√≠sticamente significativa en la conversi√≥n entre los grupos (datos filtrados).")
else:
    print("‚ùå No hay una diferencia estad√≠sticamente significativa en la conversi√≥n entre los grupos (datos filtrados).")
```

No hay una diferencia estad√≠sticamente significativa en la tasa de conversi√≥n entre los grupos A y B luego de eliminar los outliers.
Esto significa que, a√∫n despu√©s del filtrado, los resultados se mantienen consistentes con el an√°lisis anterior:
 La ligera diferencia observada entre los grupos no es suficiente para concluir que el cambio aplicado en el grupo B haya tenido un efecto real en la conversi√≥n de los usuarios.

```python
orders_filtered = orders[orders['revenue'] <= 900]

sample_A_filtered_revenue = orders_filtered[orders_filtered['group'] == 'A']['revenue']
sample_B_filtered_revenue = orders_filtered[orders_filtered['group'] == 'B']['revenue']

stat_revenue, p_value_revenue = stats.mannwhitneyu(sample_A_filtered_revenue, sample_B_filtered_revenue)

print('Estad√≠stico:', stat_revenue)
print('p-valor:', p_value_revenue)

alpha = 0.05
if p_value_revenue < alpha:
    print("‚úÖ Hay una diferencia estad√≠sticamente significativa en el tama√±o promedio del pedido entre los grupos (datos filtrados).")
else:
    print("‚ùå No hay una diferencia estad√≠sticamente significativa en el tama√±o promedio del pedido entre los grupos (datos filtrados).")
```

Esto significa que no hay una diferencia estad√≠sticamente significativa en el tama√±o promedio del pedido entre los grupos A y B despu√©s de filtrar los datos (eliminando los outliers con precios mayores a 900).
En otras palabras:
Aunque el grupo B podr√≠a haber mostrado un promedio de pedido ligeramente mayor, esta diferencia no es suficiente desde el punto de vista estad√≠stico para concluir que el cambio aplicado haya tenido un impacto real sobre el valor promedio de los pedidos.
